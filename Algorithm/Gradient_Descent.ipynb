## 경사 하강법 (Gradient Descent)

경사 하강법은 **손실함수(Loss Function)를 최소화하기 위해 모델의 파라미터를 업데이트하는 알고리즘**이다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
w = 0
b = 0
lr = 0.01
epochs = 100
loss_h = []
n = len(x)
```

```python
for epoch in range(epochs) :
    y_pred = w * x + b
    
    # 손실 함수
    loss = np.mean((y - y_pred) ** 2)
    loss_h.append(loss)

    # 경사하강법
    dw = (-2/n) * np.sum(x * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)

    w = w - lr * dw
    b = b - lr * db

    print(epoch, loss)
    
```

```python
plt.plot(loss_h)
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```
